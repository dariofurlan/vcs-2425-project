```python
from PIL import Image
import cv2

base_path="rgbd-dataset/apple/apple_1/apple_1_1_1_"
depth = base_path + "depthcrop.png"
rgb = base_path + "crop.png"

# Load the image
depth_image_cv = cv2.imread(depth, cv2.IMREAD_ANYDEPTH)  # This corresponds to IPL_DEPTH_16U for 16-bit images
depth_image = Image.fromarray(depth_image_cv)
rgb_image = Image.open(rgb)
```

```python
# Display image mode and size
print(f"Image mode: {depth_image.mode}")
print(f"Image size: {depth_image.size}")

# Get pixel value range
pixel_values = list(depth_image.getdata())
print(f"Pixel value range: {min(pixel_values)} to {max(pixel_values)}")
```

```python
import numpy as np

import matplotlib.pyplot as plt
# Plot the depth and RGB images side by side
plt.figure(figsize=(10, 5))

# Depth image
plt.subplot(1, 2, 1)
# Normalize the depth image to range [0, 1] for visualization
depth_array = np.array(depth_image)
normalized_depth = (depth_array - depth_array.min()) / (depth_array.max() - depth_array.min())

# Create a color map
plt.imshow(normalized_depth, cmap='Spectral')
plt.title("Depth Image")

# Add a color bar to map values to colors
plt.colorbar(label="Depth Value")
plt.axis('off')

# RGB image
plt.subplot(1, 2, 2)
plt.imshow(rgb_image)
plt.title("RGB Image")
plt.axis('off')

plt.show()
```
