# Demo: ApplyColormap Transform on Depth Estimation Image
This notebook demonstrates how to apply a colormap transformation to a grayscale depth estimation image using a custom `ApplyColormap` class. The process includes loading the image, applying random data augmentation, and visualizing the results side by side.

```python
# Import required libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from vcs2425 import ApplyColormap
```

## Load the Image
Define a helper function to load an image from a given file path using PIL.
Set the path to the depth estimation image and load it into memory.

```python
# Load the image
def load_image(path):
    return Image.open(path)
```

```python
# Path to the image
img_path = 'example_16bit_depth.png'
img = load_image(img_path)
```

## Define Transformation Pipeline
Create a transformation pipeline that applies a random horizontal flip and the colormap transformation to the image.

```python
from torchvision import transforms
# Load and transform the image
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=1), # we flip the image every time
    ApplyColormap(cmap='Spectral'),
])
```

### Apply Transform
Apply the defined transformation pipeline to the loaded image to obtain a color-mapped version.

```python
img_colored = train_transform(img)

print(img_colored.shape)
```

## Visualize Results
Display the original grayscale image and the color-mapped image side by side for comparison.

```python
# Display the result
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title('Original Grayscale')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Colormapped')
plt.imshow(img_colored)
plt.axis('off')

plt.tight_layout()
plt.show()
```
