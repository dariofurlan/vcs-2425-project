#### Stacked 3-Channel Conversion

The following steps are used to convert each 16-bit single-channel depth image into a standard 8-bit, 3-channel (RGB) PNG image:

- **8-bit normalization:** The original 16-bit depth values are scaled down to the 8-bit range (0–255).
- **Channel duplication:** The single 8-bit channel is duplicated three times to create a 3-channel image, making it compatible with models expecting RGB input.
- **Saving as PNG:** The resulting 3-channel images are saved in PNG format.

All 10,000 processed images are saved into a folder and uploaded to Google Drive to make the training phase faster.

```python
import os
import numpy as np
from PIL import Image

# Define input and output directories
input_dir = 'train_images'
output_dir = 'three_channel_depth'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Get a list of all .png files in the input directory
image_files = [f for f in os.listdir(input_dir) if f.endswith('.png')]

# Process each image
for image_file in image_files:
    # Construct the full path for the input image
    input_path = os.path.join(input_dir, image_file)

    # Construct the full path for the output image
    output_path = os.path.join(output_dir, image_file)

    try:
        # --- Load the 16-bit image ---
        img = Image.open(input_path)

        # Check if the image mode is I;16 (16-bit grayscale)
        if img.mode == 'I;16':
            img_array = np.array(img)

            # --- Normalize to 8-bit ---
            # Scale the 16-bit values (0-65535) to the 8-bit range (0-255)
            depth_8bit = (img_array / 256).astype(np.uint8)

            # --- Create 3-channel image ---
            # Duplicate the single 8-bit channel three times to create a 3-channel image
            depth_3ch_8bit = np.stack([depth_8bit] * 3, axis=-1)

            # --- Convert back to PIL Image and save ---
            # Convert the NumPy array back to a PIL Image in RGB mode
            output_img = Image.fromarray(depth_3ch_8bit, mode='RGB')

            # Save the processed image to the output directory
            output_img.save(output_path)

            print(f"Processed and saved: {image_file}")

        else:
            print(f"Skipping {image_file}: Not in I;16 mode.")

    except Exception as e:
        print(f"Error processing {image_file}: {e}")

print("Processing complete.")
```

#### Additinal checking
Check that all images are not corrupted and are in the correct format.

This step attempts to open each image and verifies that it is a valid, accessible file with the expected number of channels and bit depth.  
It ensures data integrity before using the images for further processing or training.

```python
import os
from PIL import Image

# Define the directory containing the processed images
processed_images_dir = 'three_channel_8bit_depth_images'

# Get list of all .png files in the directory
image_files = [f for f in os.listdir(processed_images_dir) if f.endswith('.png')]

corrupted_images = []
incorrect_format_images = []

# Iterate through each image file
for image_file in image_files:
    img_path = os.path.join(processed_images_dir, image_file)

    try:
        # Open the image
        img = Image.open(img_path)

        # Check the image mode (channels and bit depth)
        # 'RGB' mode indicates 3 channels (Red, Green, Blue), typically 8 bits per channel
        if img.mode != 'RGB':
            incorrect_format_images.append((image_file, f"Incorrect mode: {img.mode}"))
            continue

        # Check the number of channels from the band names (another way to confirm 3 channels)
        if len(img.getbands()) != 3:
             incorrect_format_images.append((image_file, f"Expected 3 channels (RGB), but found {len(img.getbands())}"))
             continue

        # While PIL's 'RGB' mode implies 8-bit, a more direct check could involve converting to numpy
        # and checking the dtype, but for simplicity and efficiency with PIL, checking the mode is usually sufficient for 'RGB'.
        # If you need absolute certainty of 8-bit data, you could add:
        # import numpy as np
        # img_array = np.array(img)
        # if img_array.dtype != np.uint8:
        #     incorrect_format_images.append((image_file, f"Incorrect data type: {img_array.dtype}"))


    except Exception as e:
        # Catch any errors during image opening or processing
        corrupted_images.append((image_file, str(e)))

# Report the findings
print("--- Checking Image Format ---")
if corrupted_images:
    print("\nCorrupted Images Found:")
    for file, error in corrupted_images:
        print(f"- {file}: {error}")
else:
    print("\nNo corrupted images found (all images could be opened).")

if incorrect_format_images:
    print("\nImages with Incorrect Format (not 8-bit, 3-channel RGB):")
    for file, reason in incorrect_format_images:
        print(f"- {file}: {reason}")
else:
    print("\nAll opened images appear to be in the correct 8-bit, 3-channel RGB format.")

if not corrupted_images and not incorrect_format_images:
    print("\nAll images in 'three_channel_depth' directory are accessible and appear to be in the correct 8-bit, 3-channel RGB format.")
```

```python
import os
import shutil

# Define the paths
source_image_dir = 'three_channel_8bit_depth_images'  # Your source folder with processed images
labels_file = 'ground_truth_reduced.txt'  # Your text file containing image filenames and labels
output_base_dir = 'organized_3channel_8bit_depth_images'  # The base directory to create label folders

# Create the output base directory if it doesn't exist
os.makedirs(output_base_dir, exist_ok=True)

# Dictionary to store image filename and its corresponding label
image_labels = {}

# Read the labels file
try:
    with open(labels_file, 'r') as f:
        for line in f:
            # Split the line into filename and label
            parts = line.strip().split()
            if len(parts) == 2:
                filename = parts[0]
                label = parts[1]
                image_labels[filename] = label
            else:
                print(f"Skipping invalid line in labels file: {line.strip()}")
except FileNotFoundError:
    print(f"Error: Labels file '{labels_file}' not found.")
    # Exit or handle the error appropriately
    exit()
except Exception as e:
    print(f"Error reading labels file: {e}")
    exit()

# Process and organize the images
processed_count = 0
skipped_count = 0

# Get a list of all image files in the source directory
source_image_files = [f for f in os.listdir(source_image_dir) if f.endswith('.png')]

for image_file in source_image_files:
    # Check if the image has a label in the dictionary
    if image_file in image_labels:
        label = image_labels[image_file]

        # Create the output directory for this label if it doesn't exist
        label_output_dir = os.path.join(output_base_dir, label)
        os.makedirs(label_output_dir, exist_ok=True)

        # Define source and destination paths
        source_path = os.path.join(source_image_dir, image_file)
        destination_path = os.path.join(label_output_dir, image_file)

        try:
            # Copy the image to the label folder
            # Use shutil.copy() if you want to keep the original images
            # Use shutil.move() if you want to move the original images
            shutil.copy(source_path, destination_path)
            processed_count += 1
            # Optional: print progress every 1000 files
            if processed_count % 1000 == 0:
                 print(f"Processed {processed_count} images...")

        except FileNotFoundError:
            print(f"Error: Source image file not found: {source_path}")
            skipped_count += 1
        except Exception as e:
            print(f"Error copying image {image_file}: {e}")
            skipped_count += 1
    else:
        print(f"Skipping image {image_file}: No label found in {labels_file}")
        skipped_count += 1

print("\n--- Organizing Complete ---")
print(f"Total images processed: {processed_count}")
print(f"Total images skipped: {skipped_count}")
```
