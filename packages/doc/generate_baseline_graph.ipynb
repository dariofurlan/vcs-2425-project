```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
colormaps = ['stacked', 'magma', 'viridis', 'plasma']
models = ['AlexNet', 'VGG19', 'ResNet50', 'InceptionV3']

# Top-1 and Top-5 accuracy data for each model and colormap
top1 = np.array([
    [0.35, 0.58, 0.64, 0.78],  # AlexNet
    [0.45, 0.43, 0.48, 0.42],  # VGG19
    [0.69, 0.59, 0.40, 0.52],  # ResNet50
    [0.30, 0.25, 0.29, 0.26],  # InceptionV3
])

top5 = np.array([
    [2.3, 2.39, 2.78, 3.22],   # AlexNet
    [2.42, 2.44, 2.47, 2.16],  # VGG19
    [2.74, 2.95, 2.84, 2.78],  # ResNet50
    [1.97, 2.17, 2.05, 1.9],   # InceptionV3
])

# Set bar width and positions
bar_width = 0.2
x = np.arange(len(colormaps))

# Define colors for each model
base_colors = ['blue', 'green', 'red', 'purple']

# Plot
fig, ax = plt.subplots(figsize=(12, 6))

for i, model in enumerate(models):
    ax.bar(x + i * bar_width, top1[i], width=bar_width, color=base_colors[i], label=f'{model} Top-1')
    ax.bar(x + i * bar_width, top5[i] - top1[i], bottom=top1[i], width=bar_width,
           color=plt.cm.get_cmap('Blues' if i==0 else 'Greens' if i==1 else 'Reds' if i==2 else 'Purples')(0.4),
           label=f'{model} Top-5')

# Aesthetics
ax.set_xlabel('Colormap')
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy by Colormap')
ax.set_xticks(x + bar_width * 1.5)
ax.set_xticklabels(colormaps)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.set_ylim(0, max(top5.flatten()) + 0.5)

plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
